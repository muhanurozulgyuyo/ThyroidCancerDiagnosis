```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv("../data/train.csv")
df_train.head()
```

```python
train_target = df_train['Cancer']
train_target
```

```python
train_input = df_train.drop("Cancer", axis=1, inplace=False)
train_input
```

```python
train_input['Race'].unique()
```

```python
train_input['Age'].max()
```

```python
bins = [0, 19, 29, 39, 49, 59, 69, 79, 100]
labels = ['0-19', '20s', '30s', '40s', '50s', '60s', '70s', '80-']

train_input['age_group'] = pd.cut(train_input['Age'], bins=bins, labels=labels)
```

```python
train_input['Country'].unique()
```

```python
# one-hot encoding
onehot_cols = ['Gender', 'Country', 'Race', 'Family_Background', 'Radiation_History', 'Iodine_Deficiency', 'Smoke', 'Weight_Risk', 'Diabetes', 'age_group']

train_input = pd.get_dummies(train_input, columns=onehot_cols, drop_first=False)
```

```python
train_input.drop(['Age', 'ID'], axis=1, inplace=True)
```

```python
train_input.head()
```

```python
train_input.info()
```

```python
# 중요한 컬럼 선택 with Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate

rf = RandomForestClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(rf, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(train_input, train_target)
importances = rf.feature_importances_

feature_names = train_input.columns
feature_imp_df = pd.DataFrame({
    'feature': feature_names,
    'importance': importances
}).sort_values(ascending=False, by='importance')

feature_imp_df
```
