```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv("../data/train.csv")
df_train.head()
```

```python
df_train['Race'].unique()
```

```python
df_train['Age'].max()
```

```python
bins = [0, 19, 29, 39, 49, 59, 69, 79, 100]
labels = ['0-19', '20s', '30s', '40s', '50s', '60s', '70s', '80-']

df_train['age_group'] = pd.cut(df_train['Age'], bins=bins, labels=labels)
```

```python
df_train['Country'].unique()
```

```python
# one-hot encoding
onehot_cols = ['Gender', 'Country', 'Race', 'Family_Background', 'Radiation_History', 'Iodine_Deficiency', 'Smoke', 'Weight_Risk', 'Diabetes', 'age_group']

df_encoded = pd.get_dummies(df_train, columns=onehot_cols, drop_first=False)
```

```python
df_encoded.drop('Age', axis=1, inplace=True)
```
